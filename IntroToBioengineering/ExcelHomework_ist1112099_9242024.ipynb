```python
#2.1
#import sequence
with open('Homework_seq.fasta', 'r') as f:
    next(f) #skip header
    seq_str = f.read() #put it in string
    seq_str = seq_str.replace("\n", "") #remove newlines

#number of nt
print("2.1 answers:")
print("The number of nts in the sequence is: ", len(seq_str))

#number of codons
#if we define that ORF is stationary with ATG and ends with TAA then
print("The number of codons in the sequence is: ", int(len(seq_str)/3))

#2.2
print("\n2.2 answers")
print("The number of A nt is: ", seq_str.count("A"))
print("The number of C nt is: ", seq_str.count("C"))
print("The number of G nt is: ", seq_str.count("G"))
print("The number of T nt is: ", seq_str.count("T"))

#2.3
gc_content = (seq_str.count("G") + seq_str.count("C"))/len(seq_str)
print("\n2.3 answers")
print("The GC content is: ", round(gc_content, 2))
print("Because the GC content is only 0.22 and the optimal growth temperature is 35C then most likely it is Nonhalophilic archaea")

#2.4
#generate primary 5`-3` seq
print("\n2.4 answers")
print("5`-3` sequence is: ", seq_str)

#2.5
#generate complementary 3`- 5` seq
comp_str = ""
for i in range(len(seq_str)):
  if seq_str[i] == "A":
    comp_str += "T"
  elif seq_str[i] == "T":
    comp_str += "A"
  elif seq_str[i] == "G":
    comp_str += "C"
  else:
    comp_str += "G"

print("\n2.5 answers")
print("5`-3` sequence is: ", seq_str)
print("3`-5` sequence is: ", comp_str)

#2.6
#generate 5`-3` mRNA sequence
mrna_str = ""
for i in range(len(comp_str)):
  if comp_str[i] == "A":
    mrna_str += "U"
  elif comp_str[i] == "T":
    mrna_str += "A"
  elif comp_str[i] == "G":
    mrna_str += "C"
  else:
    mrna_str += "G"

print("\n2.6 answers")
print("5`-3` mRNA sequence is: ", mrna_str)
print("3`-5`  DNA sequence is: ", comp_str)

#2.7 and 2.8 (get codon table)
#import codon table from https://gist.github.com/juanfal/09d7fb53bd367742127e17284b9c47bf
codon_table = {}
with open('gistfile1.txt', 'r') as codons:
  for line in codons:
    if ":" in line:
      (key, val) = line[5:8], line[12:13] #slice out codons and corresponding aa and put them in a dict
      codon_table[key] = val
    else:
      continue

print("\n2.7 and 2.8 answers")
print("Codon table is: ")
for key, value in codon_table.items():
  print(key, value)

#2.9
amino_seq = ""
for i in range(int(len(mrna_str)/3)):
  codon = seq_str[i*3:i*3+3]
  amino_seq += codon_table[codon]

print("\n2.9 answers")
print("Amino acid sequence is: ", amino_seq)

#2.10
short_seq_str = seq_str[:seq_str.find("GGG")] + "TAA"
short_amino_seq = ""
for i in range(int(len(short_seq_str)/3)):
  codon = short_seq_str[i*3:i*3+3]
  short_amino_seq += codon_table[codon]
print("\n2.10 answers")
print("Original protein sequence is: ", amino_seq, " with the lenght of", len(amino_seq)-1, "amino acids")
print("Shorter  protein sequence is: ", short_amino_seq, " with the lenght of", len(short_amino_seq)-1, "amino acids")

#2.11
#Only cysteine and methionine contain one sulfur atom in their molecule of all traditional amino acids
import scipy.constants
print("\n2.11 answers")
print("1 mole of protein equals to ", scipy.constants.Avogadro, "protein molecules in the mixture")
print("1 protein molecule has 2 cysteines and 2 methionines, which equal to 4 sulfur atoms per molecule")
print("Therefore, the number of sulfur atoms in the mixture is ", scipy.constants.Avogadro*4)
```
